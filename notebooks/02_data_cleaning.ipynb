```python
import pandas as pd
import numpy as np

# import/load data
df = pd.read_csv(r"C:\Users\Harshit\E-Commerce -Sales_Analytics\data\raw.csv",encoding="ISO-8859-1")
```

```python
# remove duplicate
df = df.drop_duplicates()
```

```python
# before drop, df.shape = (1067371, 8)
df.shape
```

```python
# removing missing descriptions
df = df.dropna(subset=["Description"])
```

```python
# removing cancelled invoices
df = df[~df["Invoice"].astype(str).str.startswith("C")]
```

```python
# removing invalid quantities
df = df[df["Quantity"] > 0]
```

```python
# removing invalid prices
df = df[df["Price"] > 0]
```

```python
# convert into date datatype
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
```

```python
# creating revenue column
df["Revenue"] = df["Quantity"] * df["Price"]
```

```python
# checking final data
df.info()
```

```python
df.shape
```

```python
df.to_csv(r"C:\Users\Harshit\E-Commerce -Sales_Analytics\data\cleaned_retail.csv",index=False)
```

```python
# reset index
df = df.reset_index(drop=True)
```

```python
df.index
```

```python
# creating different columns for year, month, month name, day and day name
df["Year"] = df["InvoiceDate"].dt.year
df["Month"] = df["InvoiceDate"].dt.month
df["Month_Name"] = df["InvoiceDate"].dt.month_name()
df["Day"] = df["InvoiceDate"].dt.day
df["Day_Name"] = df["InvoiceDate"].dt.day_name()
```

```python
df.to_csv(r"C:\Users\Harshit\E-Commerce -Sales_Analytics\data\cleaned_retail.csv",index=False)
```
